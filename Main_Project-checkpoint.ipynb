```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\AGATHI\Downloads\spotify_churn_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# 1. Churn Rate
churn_rate = df['is_churned'].mean() * 100
print(f"Overall Churn Rate: {churn_rate:.2f}%")
```

```python
# Distribution of Subscription Types
sub_dist = df['subscription_type'].value_counts(normalize=True) * 100
print("\nSubscription Type Distribution (%):")
print(sub_dist)

plt.figure(figsize=(6,4))
sns.countplot(x='subscription_type', data=df, hue='subscription_type', palette='viridis', legend=False)
plt.title("Distribution of Subscription Types")
plt.xlabel("Subscription Type")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Listening Time Distribution
plt.figure(figsize=(6,4))
sns.histplot(df['listening_time'], bins=30, kde=True, color='green')
plt.title("Listening Time Distribution")
plt.xlabel("Listening Time")
plt.show()
```

```python
# 4. Correlation Heatmap
plt.figure(figsize=(8,6))
corr = df[['age','listening_time','songs_played_per_day','skip_rate','ads_listened_per_week','offline_listening','is_churned']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 1) Which subscription type has the highest churn rate?
sub_churn = df.groupby('subscription_type')['is_churned'].mean() * 100
print("\nChurn Rate by Subscription Type (%):")
print(sub_churn)
```

```python
# 2) Do users with higher skip_rate churn more?
skip_compare = df.groupby('is_churned')['skip_rate'].mean()
print("\nAverage Skip Rate by Churn Status:")
print(skip_compare)

plt.figure(figsize=(6,4))
sns.boxplot(x='is_churned', y='skip_rate', data=df, hue='is_churned', palette='coolwarm', legend=False)
plt.title("Skip Rate vs Churn")
plt.xticks([0,1], ["Active", "Churned"])
plt.show()
```

```python
# 3) Does offline listening reduce churn?
offline_churn = df.groupby('offline_listening')['is_churned'].mean() * 100
print("\nChurn Rate by Offline Listening (%):")
print(offline_churn)

plt.figure(figsize=(6,4))
sns.barplot(
    x='offline_listening', 
    y='is_churned', 
    data=df, 
    estimator=lambda x: sum(x)/len(x)*100, 
    hue='offline_listening',   # Fix for future warning
    palette="mako", 
    legend=False
)
plt.title("Churn Rate vs Offline Listening")
plt.xlabel("Offline Listening (0 = No, 1 = Yes)")
plt.ylabel("Churn Rate (%)")
plt.show()
```
